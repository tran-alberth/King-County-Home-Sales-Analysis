### 03. Data wrangling, cleaning, and consistency checks

```python
# dropping unnecessary columns for analysis
df = df.drop(columns = ['id'])
```

```python
df_clean = df
```

```python
# check for missing values
df_clean.isnull().sum()
```

```python
df_clean
```

```python
df_clean.head()
```

```python
# checking for duplicate values
df_dups = df[df.duplicated()]
```

```python
# return df_dups
df_dups
```

There are no duplicate values.

```python
df_clean.shape
```

```python
# checking for mixed-type data
for col in df_clean.columns.tolist():
  weird = (df_clean[[col]].applymap(type) != df_clean[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_clean[weird]) > 0:
    print (col)
```

No mixed data-types were found.

```python
# displaying data types
df_clean.info()
```

No changes to data type need to be done.

```python
# Converting 'data' from 'object' to 'datetime'
df_clean['date'] = pd.to_datetime(df_clean['date'])
```

```python
# changing data format from yyyy-mm-dd to yyyy-mm
df_clean['date'] = df_clean['date'].dt.strftime('%Y-%m')
```
